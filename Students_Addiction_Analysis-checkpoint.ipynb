## K-Means Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Students Social Media Addiction.csv")

# Clean column names
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('-', '_')
print("Cleaned Columns:", df.columns.tolist())

# Select meaningful numeric features
features = ['Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night', 'Mental_Health_Score', 'Addicted_Score']

# Verify they exist
missing = [f for f in features if f not in df.columns]
if missing:
    raise KeyError(f"Missing columns in dataset: {missing}")

# Prepare data
X = df[features].dropna().values

# Initialize K-Means parameters
k = 3
max_iters = 100
np.random.seed(42)

# Randomly choose initial centroids
centroids = X[np.random.choice(range(len(X)), k, replace=False)]

# Run K-Means manually
for i in range(max_iters):
    distances = np.sqrt(((X - centroids[:, np.newaxis])**2).sum(axis=2))
    labels = np.argmin(distances, axis=0)
    new_centroids = np.array([X[labels == j].mean(axis=0) for j in range(k)])
    if np.allclose(centroids, new_centroids):
        break
    centroids = new_centroids

# Add cluster labels to dataframe
df = df.loc[df[features].notna().all(axis=1)].copy()
df['Cluster'] = labels

# Visualize (2D example)
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=200, marker='X')
plt.title("K-Means Clustering of Students' Social Media Addiction (No sklearn)")
plt.xlabel(features[0])
plt.ylabel(features[1])
plt.show()

# Display cluster summaries
for i in range(k):
    cluster_data = df[df['Cluster'] == i]
    print(f"\nCluster {i} Summary:")
    print(cluster_data[features].mean().round(2))
```

## Addiction Intensity Score

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv("Students Social Media Addiction.csv")

# Clean column names (optional but good practice)
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('-', '_')
print("Cleaned Columns:", df.columns.tolist())

# Create a custom 'Addiction_Score' using available features
df['Addiction_Score'] = (
    (df['Avg_Daily_Usage_Hours'] * 0.4) + 
    (df['Mental_Health_Score'] * 0.3) - 
    (df['Sleep_Hours_Per_Night'] * 0.3)
)

# Normalize score between 0‚Äì100
df['Addiction_Score'] = 100 * (df['Addiction_Score'] - df['Addiction_Score'].min()) / (df['Addiction_Score'].max() - df['Addiction_Score'].min())

# If no 'Student_Name' column exists, use 'Student_ID' instead
if 'Student_Name' not in df.columns:
    df['Student_Name'] = df['Student_ID']

# Top 10 most addicted students
top10 = df.nlargest(10, 'Addiction_Score')

# Plot
plt.figure(figsize=(8,5))
plt.barh(top10['Student_Name'], top10['Addiction_Score'], color='tomato')
plt.title("üî• Top 10 Most Addicted Students (Custom Score)")
plt.xlabel("Addiction Score (0‚Äì100)")
plt.gca().invert_yaxis()
plt.show()

# Print summary
print("\nTop 10 Most Addicted Students:")
print(top10[['Student_Name', 'Addiction_Score']])
```
